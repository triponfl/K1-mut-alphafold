```python
%pip -q install -U nglview ipywidgets jupyterlab_widgets biopython

from pathlib import Path
import zipfile

import nglview as nv
from Bio.PDB.MMCIFParser import MMCIFParser
from Bio.PDB import PDBParser
```

```python
#can be ignored
files_dir = Path("Files")
out_root = Path("unzipped")
out_root.mkdir(exist_ok=True)

zips = sorted(files_dir.glob("fold_*.zip"))
print("Found zip files:", len(zips))

for z in zips:
    sub = out_root / z.stem
    sub.mkdir(exist_ok=True)
    with zipfile.ZipFile(z) as zz:
        zz.extractall(sub)

print("Unzipped into:", out_root.resolve())
```

```python
def find_model_file(folder: Path):
    cifs = sorted(folder.rglob("*.cif")) + sorted(folder.rglob("*.mmcif"))
    pdbs = sorted(folder.rglob("*.pdb")) #look over - i don't think there are any
    if cifs: return cifs[0]
    if pdbs: return pdbs[0]
    return None

models = {}
for sub in sorted(Path("unzipped").iterdir()):
    if sub.is_dir():
        f = find_model_file(sub)
        if f:
            models[sub.name] = f

print("Models found:", len(models))
models.keys()
```

```python
def find_model_file(folder: Path):
    cifs = sorted(folder.rglob("*.cif")) + sorted(folder.rglob("*.mmcif"))
    pdbs = sorted(folder.rglob("*.pdb"))
    if cifs: return cifs[0]
    if pdbs: return pdbs[0]
    return None

models = {}
for sub in sorted(Path("unzipped").iterdir()):
    if sub.is_dir():
        f = find_model_file(sub)
        if f:
            models[sub.name] = f

print("Models found:", len(models))
print("Keys:", list(models.keys()))
```

```python
def show_model(name):
    path = models[name]
    if path.suffix.lower() == ".pdb":
        structure = PDBParser(QUIET=True).get_structure(name, str(path))
    else:
        structure = MMCIFParser(QUIET=True).get_structure(name, str(path))

    view = nv.show_biopython(structure)
    view.clear_representations()
    view.add_cartoon(color="bfactor")
    return view
```
